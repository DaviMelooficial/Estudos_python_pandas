```python
import pandas as pd
```

```python
# Data para o DataFrame
data = {'Classe': ["Júnior","Júnior","Pleno","Pleno","Sênior","Sênior"],
        'Nome': ["Jorge","Carlos","Roberto","Rafael","Athur","Silas"],
        'Venda': [200, 120, 340, 124, 243, 350]}
```

```python
# Definição do DataFrame
df = pd.DataFrame(data)
df
```

```python
grupo = df.groupby("Classe") # Função de agrupamento por coluna
```

```python
grupo.sum(numeric_only=True) # Função de soma utilizando somente números
```

```python
grupo.mean(numeric_only=True) # Função de média
```

```python
grupo.min() # Função de mínimo
```

```python
grupo.max() # Função de máximo
```

```python
df_2 = df.copy() # Atribui ao df_2 uma cópia do df
```

```python
df_2["Venda"] = [120,380,144,225,180,600] # Mudança dos valores da coluna ['Venda']
```

```python
df_2
```

```python
df_3 = pd.concat([df,df_2]) # Concatenação dos dois DataFrames e atribuindo a um novo
```

```python
df_3
```

```python
df_3.groupby(["Classe", "Nome"]).sum() # Agrupamento por ['Classe'] e ['Nome'], em seguida somando os valores de cada grupo
```
